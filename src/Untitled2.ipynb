```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("pns2019.csv")


data = data[['V0001', 'Q00201', 'Q00202', 'C008', 'Q00401', 'Q00502', 'Q026', 'Q028', 'C006']]
data.columns = ['Estado', 'Diagnóstico_Hipertensao', 'Hipertensao_Periodo_Gravidez', 'Idade', 'Visita_Medico_Regulares', 'Motivo_Nao_Visitar_Medico', 'Internacao_Hipertensao', 'Grau_Limitacao_Atividades', 'Sexo']


codigo_regiao_map = {
    
    35: 'Sudeste',  # São Paulo
    33: 'Sudeste',  # Rio de Janeiro
    31: 'Sudeste',  # Minas Gerais
    32: 'Sudeste',  # Espírito Santo
 
    29: 'Nordeste',  # Bahia
    28: 'Nordeste',  # Sergipe
    27: 'Nordeste',  # Alagoas
    26: 'Nordeste',  # Pernambuco
    25: 'Nordeste',  # Paraíba
    24: 'Nordeste',  # Rio Grande do Norte
    23: 'Nordeste',  # Ceará
    22: 'Nordeste',  # Piauí
    21: 'Nordeste',  # Maranhão
  
    43: 'Sul',  # Rio Grande do Sul
    42: 'Sul',  # Santa Catarina
    41: 'Sul',  # Paraná

    16: 'Norte',  # Amapá
    15: 'Norte',  # Pará
    14: 'Norte',  # Roraima
    13: 'Norte',  # Amazonas
    12: 'Norte',  # Acre
    11: 'Norte',  # Rondônia
    17: 'Norte',  # Tocantins

    53: 'Centro-Oeste',  # Distrito Federal
    52: 'Centro-Oeste',  # Goiás
    51: 'Centro-Oeste',  # Mato Grosso
    50: 'Centro-Oeste',  # Mato Grosso do Sul
}


regioes = list(set(codigo_regiao_map.values()))


for regiao in regioes:
    
    dados_regiao = data[data['Estado'].map(codigo_regiao_map) == regiao]
    

    contagem_diagnostico = dados_regiao['Diagnóstico_Hipertensao'].value_counts()
    
  
    contagem_sexo = dados_regiao['Sexo'].value_counts()

    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
   
    axs[0].pie(contagem_diagnostico, labels=contagem_diagnostico.index, autopct='%1.1f%%', startangle=90)
    axs[0].set_title(f'Distribuição de Diagnóstico de Hipertensão na Região {regiao}')
    
   
    axs[1].pie(contagem_sexo, labels=contagem_sexo.index, autopct='%1.1f%%', startangle=90)
    axs[1].set_title(f'Distribuição por Sexo na Região {regiao}')
    
   
    plt.tight_layout()
    

    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hipertensao_31_40 = hipertensao[(hipertensao['Q003'] >= 31) & (hipertensao['Q003'] <= 40)]
```

```python
idade_cor_raca = hipertensao_31_40[['Q003', 'C009']].dropna()
```

```python
mapa_cor = {
    1: 'Branca',
    2: 'Preta',
    3: 'Amarela',
    4: 'Parda',
    5: 'Indígena'
}
idade_cor_raca['Cor ou Raça'] = idade_cor_raca['C009'].map(mapa_cor)
```

```python
idade_cor_raca.rename(columns={'Q003': 'Idade'}, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=idade_cor_raca, x='Cor ou Raça', y='Idade')
plt.xlabel('Cor ou Raça')
plt.ylabel('Idade')
plt.title('Distribuição da Idade em Relação à Cor ou Raça (31-40 anos)')
plt.xticks(rotation=45)
plt.show()
```

```python
contagens_por_cor_raca = idade_cor_raca['Cor ou Raça'].value_counts()
```

```python
for cor_raca, contagem in contagens_por_cor_raca.items():
    print(f"Cor ou Raça '{cor_raca}': {contagem} registros")
```
